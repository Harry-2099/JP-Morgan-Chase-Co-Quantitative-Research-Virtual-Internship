```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Task 3 and 4_Loan_Data.csv')
df.head()
```

```python
# function that gets the log_likelihood after binning using cut
def Log_likelihood_bins(bins,data):
    data['fico_bins'] = pd.cut(data['fico_score'], bins = bins, labels = [i for i in range(1,bins+1)]) # assumes a column called fico_score
    log_likelihood = 0 
    for i in range(1,bins+1):
        n = len(data.loc[data['fico_bins']==i])
        k = sum(data.loc[data['fico_bins']==i,"default"])
        p = k/n if n > 0 else 0  # ensuring no division by 0
        if p == 0 or p == 1: # both of these cases will result in a log(0) which is undefined
            ll = 0
        else:
         ll = k * np.log(p)+(n - k)* (np.log(1-p))
        log_likelihood +=ll
    return log_likelihood
Log_likelihood_bins(10, data = df)
```

```python
#making an opitmizer to get the best number of bins that maxes the Log likelihood
def optimize_bins(data,min_bins,max_bins):
    best_likelihood = float("-inf")# sets equal to -infinity so all trials will be better 
    best_bins = min_bins #setting the initial bins
    for nbins in range(min_bins,max_bins+1):
       current_likelihood= Log_likelihood_bins(nbins,data) #computes the likelihood for each iteration in the range
       if current_likelihood> best_likelihood: # changes the log LL and the number of bins for every better iteration until max_bins is reached
           best_likelihood = current_likelihood
           best_bins = nbins
    return best_bins,best_likelihood
```

```python
print(optimize_bins(df,1,20))
```
